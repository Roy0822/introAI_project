```python
import dataset
import model
import detection
import matplotlib.pyplot as plt
import time
```

### Part 1
Implement loadImages function in dataset.py and test the following code

```python
print('Loading images')
train_data = dataset.load_images('data/train')
print(f'The number of training samples loaded: {len(train_data)}')
test_data = dataset.load_images('data/test')
print(f'The number of test samples loaded: {len(test_data)}')

print('Show the first and last images of training dataset')
fig, ax = plt.subplots(1, 2)
ax[0].axis('off')
ax[0].set_title('Car')
ax[0].imshow(train_data[1][0], cmap='gray')
ax[1].axis('off')
ax[1].set_title('Non car')
ax[1].imshow(train_data[-1][0], cmap='gray')
plt.show()
```

### Part 2
Build and train 3 kinds of classifiers: KNN, Random Forest and Adaboost.

```python
car_clf = model.CarClassifier(
    model_name="AB", # KNN, RF (Random Forest) or AB (AdaBoost)
    train_data=train_data,
    test_data=test_data
    
)
start = time.process_time()


car_clf.train()
car_clf.eval()

for i in range(10000):
    "-".join(str(n) for n in range(100))
end = time.process_time()
print("執行時間：%f 秒" % (end - start))
```

```python
import numpy as np
truth_stat = []
with open('GroundTruth.txt') as file:
    for line in file.readlines():
        total = 0
        temparray = line.strip().split(' ')
        #print(temparray)
        for i in range(len(temparray)):
            if temparray[i] == '1':
                total = total + 1
        truth_stat.append(total)

AB_stat = []
with open('(AB)ML_Models_pred.txt') as file:
     for line in file.readlines():
        total = 0
        temparray = line.strip().split(' ')
        #print(temparray)
        for i in range(len(temparray)):
            if temparray[i] == '1':
                total = total + 1
        AB_stat.append(total)

RF_stat = []
with open('(RF)ML_Models_pred.txt') as file:
     for line in file.readlines():
        total = 0
        temparray = line.strip().split(' ')
        #print(temparray)
        for i in range(len(temparray)):
            if temparray[i] == '1':
                total = total + 1
        RF_stat.append(total)
        
KNN_stat = []
with open('(KNN)ML_Models_pred.txt') as file:
     for line in file.readlines():
        total = 0
        temparray = line.strip().split(' ')
        #print(temparray)
        for i in range(len(temparray)):
            if temparray[i] == '1':
                total = total + 1
        KNN_stat.append(total)
'''        
print(AB_stat)
print(RF_stat)
print(KNN_stat)
print(truth_stat)
'''

        
#generation of PSO line graph:

graph_pso,ax = plt.subplots()
#for graph useage
x_ax = [0,10,20,30,40,50]
y_ax = [40,50,60,70,80]
x=list(range(0,50))
#print(x)
ax.set_title('Parking Slot Occupation')
ax.set_xlim(0,50)    # 設定 x 軸邊界為 0～50
ax.set_ylim(40,70)    # 設定 y 軸邊界為 0～76
ax.plot(x,RF_stat, linewidth = 1,label = 'Random Forest')
ax.plot(x,KNN_stat, linewidth = 1, label = 'KNN')
ax.plot(x,AB_stat, linewidth = 1,label = 'AdaBoost')
ax.plot(x,truth_stat, linewidth = 1,label = 'Ground Truth')
ax.grid()
ax.set_xlabel('Frames',{'fontsize':10,'color':'black'})    # 設定 x 軸標籤
ax.set_ylabel('Car Detected',{'fontsize':10,'color':'black'})  # 設定 y 軸標籤
ax.legend(
    loc='best',
    fontsize=7.5,
    shadow=False,
    edgecolor='#000',
    title='Legend',
    title_fontsize=10
)

graph_pso.savefig('PSO.png',
            transparent=True,
            bbox_inches='tight',
)

#-----------------------------------------------------------------------
#generation of accuracy line graph
'''
Accuracy:
1 - (abs(sample-truth) / truth)
'''
AB_np = np.array(AB_stat)
truth_np = np.array(truth_stat)
KNN_np = np.array(KNN_stat)
RF_np = np.array(RF_stat)

relative_error = np.abs(AB_np - truth_np) / truth_np
AB_acc= 1-relative_error
relative_error = np.abs(KNN_np - truth_np) / truth_np
KNN_acc= 1-relative_error
relative_error = np.abs(RF_np - truth_np) / truth_np
RF_acc= 1-relative_error


graph_ac,axi = plt.subplots()
#for graph usage
axi.set_title('Accuracy of Models')
axi.set_xlim(0,50)    # 設定 x 軸邊界為 0～50
axi.set_ylim(0.75,1.01)    # 設定 y 軸邊界為 0～76
axi.plot(x,RF_acc, linewidth = 1,label = 'Random Forest')
axi.plot(x,KNN_acc, linewidth = 1, label = 'KNN')
axi.plot(x,AB_acc, linewidth = 1,label = 'AdaBoost')
axi.grid()
axi.set_xlabel('Frames',{'fontsize':10,'color':'black'})    # 設定 x 軸標籤
axi.set_ylabel('Accuracy',{'fontsize':10,'color':'black'})  # 設定 y 軸標籤
axi.legend(
    loc='best',
    fontsize=7.5,
    shadow=False,
    edgecolor='#000',
    title='Legend',
    title_fontsize=10
)

graph_ac.savefig('ACcC.png',
            transparent=True,
            bbox_inches='tight',
)

```
